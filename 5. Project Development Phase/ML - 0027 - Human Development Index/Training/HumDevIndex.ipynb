## Epic 4: Data Preprocessing and Label Encoding

```python
for i, col in enumerate(df.columns):
    print(i, col)
```

```python
# Independent variables (X) and dependent variable (Y)
X = df.iloc[:, [0, 1, 2, 3, 4]]   # Country, Life_Expectancy, Mean_Years_of_Schooling, Expected_Years_of_Schooling, GNI_Per_Capita
Y = df.iloc[:, 5]                 # HDI_Score
X.head()
```

```python
X.isnull().sum()
```

```python
X = X.fillna(X.mean(numeric_only=True))
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X['Country'] = le.fit_transform(X['Country'])
X.head()
```

## Epic 5: Dividing the Dataset into Train and Test Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('Y_train shape:', Y_train.shape)
print('Y_test shape:', Y_test.shape)
```

## Epic 6: Fitting the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y_test, y_pred)
print('R-squared value:', r2)
```

```python
# Test with a single row from X_test
single_prediction = model.predict(X_test.iloc[[0]])
print('Predicted HDI:', single_prediction)
print('Actual HDI:', Y_test.iloc[0])
```
